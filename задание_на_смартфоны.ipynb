```python
import pandas as pd
import matplotlib.pyplot as plt
import folium

xls = pd.ExcelFile('Weekly_sales_simulation.xlsx')
df = pd.read_excel(xls, 0)
df
```

**Показать таблицу и график с общими продажами всех брендов за весь период.**

```python
total_sales = df.groupby('Unnamed: 0')[['Week 1','Week 2','Week 3','Week 4','Week 5','Week 6','Week 7','Week 8','Week 9','Week 10','Week 11', 'Week 12', 'Week 13', 'Week 14']].sum()
total_sales
```

```python
total_sales['Week 1'] = pd.to_numeric(total_sales['Week 1'], errors='coerce')
total_sales['Week 2'] = pd.to_numeric(total_sales['Week 2'], errors='coerce')
total_sales['Week 3'] = pd.to_numeric(total_sales['Week 3'], errors='coerce')
total_sales['Week 4'] = pd.to_numeric(total_sales['Week 4'], errors='coerce')
total_sales['Week 5'] = pd.to_numeric(total_sales['Week 5'], errors='coerce')
total_sales['Week 6'] = pd.to_numeric(total_sales['Week 6'], errors='coerce')
total_sales['Week 7'] = pd.to_numeric(total_sales['Week 7'], errors='coerce')
total_sales['Week 8'] = pd.to_numeric(total_sales['Week 8'], errors='coerce')
total_sales['Week 9'] = pd.to_numeric(total_sales['Week 9'], errors='coerce')
total_sales['Week 10'] = pd.to_numeric(total_sales['Week 10'], errors='coerce')
total_sales['Week 11'] = pd.to_numeric(total_sales['Week 11'], errors='coerce')
total_sales['Week 12'] = pd.to_numeric(total_sales['Week 12'], errors='coerce')
total_sales['Week 13'] = pd.to_numeric(total_sales['Week 13'], errors='coerce')
total_sales['Week 14'] = pd.to_numeric(total_sales['Week 14'], errors='coerce')

total_sales['sum'] = total_sales[['Week 1','Week 2','Week 3','Week 4','Week 5','Week 6','Week 7','Week 8','Week 9','Week 10','Week 11', 'Week 12', 'Week 13', 'Week 14']].sum(axis=1)
total_sales
```

```python
total_sales.reset_index().plot(kind='bar', x='Unnamed: 0', y='sum')
plt.title('Total Sales')
plt.xlabel('Brand')
plt.ylabel('Sales')
plt.show()


#на первом месте среди самых продаваемых смартфонов находится - Самсунг, а на втором - Эппл. так как они самые известные и качественные марки.
#тем временем, кол-во продаваемых не всем известных смартфонов маленькое.
```

**Показать топ 10 проданных моделей по размеру дисплея + кол-во + прибыль:**

```python
models = df.pivot_table(index='Unnamed: 0', values='Unnamed: 1', aggfunc='count')
models.drop('Brand', axis=0, inplace=True)
models
```

```python
df['Unnamed: 10'] = pd.to_numeric(df['Unnamed: 10'], errors='coerce') #он превратит все значения, которые не удалось преобразовать в числа, в NaN

top_10_display = df.groupby('Unnamed: 0')['Unnamed: 10'].sum().sort_values(ascending=False).head(11)
top_10_display
```

```python
df['Unnamed: 11'] = pd.to_numeric(df['Unnamed: 11'], errors='coerce')

top_10_prof = df.groupby('Unnamed: 0')['Unnamed: 11'].sum().sort_values(ascending=False).head(11)
top_10_prof
```

```python
top_10_display = top_10_display.to_frame()
top_10_prof = top_10_prof.to_frame()
```

```python
#создав три датафрейма, соединяю в одну.

merged_df = models.merge(top_10_display, on='Unnamed: 0').merge(top_10_prof, on='Unnamed: 0')
merged_df
```

```python
merged_df.columns = ['count', 'display', 'profit']
merged_df
```

```python
plot = merged_df.plot.pie(subplots=True, figsize=(20, 20), autopct='%1.1f%%')
plt.title('Top 10 Phones')
plt
```

**Показать топ 10 проданных моделей по объему памяти + кол-во + прибыль**

```python
df['Unnamed: 3'] = pd.to_numeric(df['Unnamed: 3'], errors='coerce')

top_10_storage = df.groupby('Unnamed: 0')['Unnamed: 3'].sum().sort_values(ascending=False).head(11)
top_10_storage
```

```python
top_10_storage = top_10_storage.to_frame()
```

```python
merged_df = models.merge(top_10_storage, on='Unnamed: 0').merge(top_10_prof, on='Unnamed: 0')
merged_df
```

```python
merged_df.columns = ['count', 'storage', 'profit']
merged_df
```

```python
merged_df['storage'] = merged_df['storage'] / 100
merged_df
```

```python
merged_df['profit'] = merged_df['profit'] / 10
merged_df
```

```python
plt.plot(merged_df.index, merged_df['storage'])
plt.plot(merged_df.index, merged_df['profit'])
plt.plot(merged_df.index, merged_df['count'])
plt.legend(['Storage', 'Profit', 'Count'])
plt.title('Top 10 Phones')
plt.xlabel('Phone')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt

#воспользовалась делением, ибо число большое и анализировать сложновато.
#топ 1 проданный модель по объему памяти + кол-во + прибыль -эппл. а затем только самсунг.
#так как памяти у эппл больше, чем у сотальных смартфонов
```
